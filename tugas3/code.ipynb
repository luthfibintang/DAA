**Latihan 10** : Buatlah matrix dengan nilai berikut :

[ [100 200 300]

  [700 600 500]

  [900 1000 800]]

Lakukan Transpose matrix!

```python
import numpy as np
matrix = np.array([
    [100,200,300],
    [700,600,500],
    [900, 1000, 800]
])

transpose_matrix = np.transpose(matrix)

print("Matrix :\n", matrix , "\n")
print("Transpose Matrix :\n", transpose_matrix)
```
