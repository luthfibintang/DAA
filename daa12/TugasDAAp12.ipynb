####Pengenalan Teori *Network Analysis*

Teori analisis jaringan memungkinkan representasi data yang saling berhubungan sebagai jaringan, di mana simpul-simpul adalah unit dasar. Jaringan ini terdiri dari simpul-simpul yang saling terhubung melalui koneksi yang mewakili hubungan antar entitas. Penting untuk mengevaluasi kegunaan dan pentingnya setiap simpul dalam konteks permasalahan yang dihadapi.

---

####Shortest Path

Path adalah urutan node antara node awal dan node akhir, di mana tidak ada node yang muncul dua kali pada jalur tersebut. Jalur mewakili rute antara titik awal dan akhir yang dipilih. Ini akan menjadi himpunan simpul, p, menghubungkan simpul awal dengan simpul akhir.

Length of path dihitung dengan menghitung sisi-sisi penyusunnya. Dari semua opsi, jalur dengan panjang terkecil disebut jalur terpendek. Penghitungan jalur terpendek digunakan secara luas dalam algoritme teori graf, tetapi tidak selalu mudah untuk dihitung. Ada berbagai algoritme yang dapat digunakan untuk menemukan jalur terpendek antara node awal dan node akhir. Salah satu algoritme paling populer yang dapat digunakan untuk menemukan jarak terpendek antara simpul awal dan simpul akhir adalah algoritme Dijkstra

Membuat Neighborhood: Menemukan strategi untuk membuat lingkungan di sekitar simpul yang diinginkan sangat penting untuk algoritme grafik. Metodologi untuk menciptakan lingkungan didasarkan pada pemilihan asosiasi langsung dengan simpul yang diinginkan.

Triangles:  Dalam teori graf, menemukan simpul yang terhubung dengan baik satu sama lain penting untuk tujuan analisis. Salah satu tekniknya adalah mencoba mengidentifikasi segitiga, yang merupakan subgraf yang terdiri dari tiga simpul yang terhubung langsung satu sama lain, dalam jaringan.

Density: Kami menamakan graf di mana setiap simpul terhubung langsung ke setiap simpul lainnya sebagai jaringan yang terhubung penuh.

###Pengukurna Centrality

Ada ukuran yang berbeda untuk memahami sentralitas simpul tertentu dalam graf atau subgraf. Ukuran sentralitas berikut ini banyak digunakan dalam analisis grafik:

1. Degree
2. Betweenness
3. Closeness
4. Eigenvector

```python
# Pengukuran Centrality : Eigenvector

import networkx as nx
import matplotlib.pyplot as plt
vertices = range(1,10)
edges = [(7, 2), (2, 3), (4,5), (7,3), (7,5), (1,6), (1,7), (2,8), (2,9)]
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
#Pengukuran Degree Centrality
nx.degree_centrality(G)
```

```python
#Pengukuran Betweeness Centrality
nx.betweenness_centrality(G)
```

```python
#Pengukuran Closeness Centrality
nx.closeness_centrality(G)
```

```python
#Pengukuran EigenVector Centrality
centrality = nx.eigenvector_centrality(G)
sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
```

```python
# Pengukuran Centrality : Eigenvector

vertices =  ['Bob', 'Mike', 'Emma', 'Bob', 'Jill', 'John', 'Leah', 'Shane', 'Liz', 'Allen', 'Lisa']
edges = [('Bob', 'Mike'), ('Mike', 'Emma'), ('Bob', 'Emma'), ('John', 'Jill'),
         ('John', 'Shane'), ('John', 'Leah'), ('Leah', 'Shane'), ('Bob', 'Jill'),
         ('Jill', 'Emma'), ('Jill', 'Shane'), ('Emma', 'Liz'), ('Shane', 'Liz'),
         ('Liz', 'Allen'), ('Allen', 'Lisa')]
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
eigenvector_centrality = nx.eigenvector_centrality_numpy(G)
print("Eigenvector Centrality:")
for node, centrality in eigenvector_centrality.items():
    print(f"{node}: {centrality}")
```
