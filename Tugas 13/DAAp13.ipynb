####Implementasi Algoritma Breadth-first Search

```python
def bfs(graph, start):
  visited = []
  queue = []
  queue = [start]
  while queue:
    node = queue.pop(0)
    if node not in visited:
      visited.append(node)
      neighbours = graph[node]
      for neighbour in neighbours:
        queue.append(neighbour)
  return visited
```

```python
#Finding the Shorted Path on unweight graph
graph = {0: [9, 7, 11],
         1: [10, 8],
         2: [12, 3],
         3: [2, 4],
         4: [3],
         5: [6],
         6: [5, 7],
         7: [0, 11, 6, 3],
         8: [1, 9, 12],
         9: [0, 8, 10],
         10: [1, 9],
         11: [0, 7],
         12: [8, 2]
         }

bfs(graph, 0)
```

####Depth First Search Algorithm

```python
def dfs(visited, graph, node):
  if node not in visited:
    print(node)
    visited.add(node)
    for neighbor in graph[node]:
      dfs(visited, graph, neighbor)
```

####Analisa Kesimpulan

Lakukan Analisa terhadap algoritma tersebut BFS dan DFS. Bandingkan hasil dengan 3 data graph (Amin, Rektor, Titik 0) Kesimpulannya apa?

Kesimpulan yang dapat saya ambil dari ketiga data tersebut adalah :    
1. BFS fokus kepada pecarian yang dimualid ari simpul awal dan melebar sementara DFS fokus kepada eksl=plorasi secara dalam (depth first) maksudnya adlaah DFS akan mencari simpul sedalam mungkin pada setiap jalur yang unik secara berurutan
2. BFS menggunakan antrian (queue) untuk mencari simpul, sementara DFS menggunakan stack seperti yang dijelaskan di ppt atua bisa juga menggunakan rekursif.
